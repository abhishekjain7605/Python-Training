```python
passwd = input("Enter your password: ")
contains_upper=False
contains_lower=False
contains_digit=False
contains_special=False
special_characters = "!@#$%^&*()-+={}[]|:;'<>,.?/~`"
length = len(passwd)
def check_passwd(passwd):
    for char in passwd:
        if length<8:
            print("Password must be at least 8 characters long.")
            break
        elif char.isupper():
            contains_upper=True
        elif char.islower():
            contains_lower=True
        elif char.isdigit():
            contains_digit=True
        elif char in special_characters:
            contains_special=True
    if contains_upper and contains_lower and contains_digit and contains_special and length>=8:
        print("Strong Password")
    else:
        print("Weak Password")  
```
