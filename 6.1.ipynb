```python
import numpy as np
from scipy import stats
```

```python
data = [22,20,19,24,23,21]
# claimed mean
mu_0 = 20 
```

```python
def func(data,mu_0):
    alpha = 0.05
    t_stat , p_value = stats.ttest_1samp(data, mu_0)
    if p_value < alpha:
        print("Significant difference")
    else:
        print("no significant difference")
    print(f"t_stat {t_stat} and p_value {p_value}")
```

```python
func(data , mu_0)
```

Independent t test



```python
group1 = [10,12,14,16,18]
group2 = [28,29,11,22,23]
```

```python
t_value , p_value = stats.ttest_ind(group1,group2)
alpha = 0.05
```

```python
if p_value < alpha:
    print("Significant difference")
else:
    print("no significant difference")
```
