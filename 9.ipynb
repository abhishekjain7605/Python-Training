```python
import pandas as pd
```

```python
df = pd.read_csv('data_cleaning_practice_dataset.csv')
```

```python
df.info()
```

```python
df['income']
```

```python
df['income'] = df['income'].abs()
```

```python
df['income'] = df['income'].fillna(df['income'].mean())
```

```python
df['age']
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
age_mean = df.loc[(df['age'] > 20) & (df['age'] < 55), 'age'].mean()
```

```python
df['age'] = df['age'].fillna(age_mean)
```

```python
df.loc[df['age'] < 20, 'age'] = age_mean
```

```python
df.loc[df['age'] > 55, 'age'] = age_mean
```

```python
df['age'] = df['age'].round().astype(int)
```
