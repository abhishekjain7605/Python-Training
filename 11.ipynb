```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'name': ['rahul', 'sachin', 'virat'],
    'age': [30, 40, 35]
})

df2 = pd.DataFrame({
    'name': ['sheela', 'neha', 'priya'],
    'age': [25, 30, 28]
})
```

```python
pd.concat([df1, df2]).reset_index(drop=True)
```

```python
df3 = pd.concat([df1, df2], axis=1).reset_index(drop=True)
df3
```

```python
student = pd.DataFrame({
    'roll no': [1, 2, 3,4,5],
    'name': ['rahul', 'sachin', 'virat', 'sheela', 'neha']
})
```

```python
marks = pd.DataFrame({
    'roll no': [1, 2, 3, 4, 7, 6],
    'marks': [80, 90, 85, 95, 70, 75]
})
```

```python
pd.merge(student, marks, on='roll no')
```

```python
pd.merge(student, marks, on='roll no', how='left')
```

```python
pd.merge(student, marks, on='roll no', how='right')
```

```python
pd.merge(student, marks, on='roll no', how='outer')
```
