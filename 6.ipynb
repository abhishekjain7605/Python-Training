```python
import numpy as np
from scipy import stats
```

```python
data = [22,20,19,24,23,21]
# claimed mean
mu_0 = 20 
```

```python
n = len(data)
mean = np.mean(data)
std = np.std (data, ddof=1)
```

```python
print(f"Number of elements {n} , mean: {mean}, standard deviation {std}")
```

```python
t_value = (mean-mu_0)/(std/np.sqrt(n))
p_value = 2*(1-stats.t.cdf(abs(t_value) , df = n-1))
```

```python
print(f"t_value is {t_value} and p_value is {p_value}")
```

```python
confidence = 0.95
t_critical = stats.t.ppf((1+confidence)/2,df=n-1)
margin_of_error = t_critical*(std/np.sqrt(n))
```

```python
print(t_critical ,margin_of_error)
```

```python
lower = mean - margin_of_error
upper = mean + margin_of_error
```

```python
print(f"{lower} : {upper}")
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("Significant difference")
else:
    print("no significant difference")
```

```python
print(p_value, alpha)
```
